# Validering av inferensresultat genom jämförelse med liknande objekt i datasetet

```python
import pandas as pd

df = pd.read_csv("data/partille_sold_eda.csv")

# Försök matcha så nära din inference-input som möjligt
comp = df[
    (df["type_Villa"] == 1) &
    (df["living_area_raw"].between(85, 115)) &
    (df["plot_area_raw"].between(800, 1300)) &
    (df["house_age"].between(45, 75)) &
    (df["dist_ostra_km"].between(0, 2.0)) &      # du hade 0.8 km
    (df["dist_gbg_center_km"].between(4.5, 7.0)) # du hade ~5.5
].copy()

comp[["sold_price_raw","living_area_raw","plot_area_raw","house_age"]].describe()
comp["sold_price_raw"].median(), comp.shape[0]
```
